```python
#import libraries
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import imageio

%matplotlib inline
```

```python
#read the image
image = mpimg.imread('test.jpg')
```

```python
imgplot = plt.imshow(image)
```

```python
# grayscale image
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
edges = cv2.Canny(gray, 50, 150)
plt.imshow(edges, cmap='Greys_r')
```

```python
mask_pts = np.array([[[0,550], [475,270], [960, 550], [960,0], [0,0]]])

temp_img = np.copy(edges)
cv2.fillPoly(temp_img, mask_pts, [0,0,0])
plt.imshow(temp_img)
```

```python
prev = np.copy(image)
cv2.polylines(prev, mask_pts, True, (0,255,255),10)
plt.imshow(prev)
```

```python
rho = 1
theta = np.pi / 180
threshold = 30
min_line_length = 70
max_line_gap = 5
line_image = np.copy(image)*0

lines = cv2.HoughLinesP(temp_img, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)
for line in lines:
    for x1, y1, x2, y2 in line:
        cv2.line(line_image, (x1,y1),(x2,y2), (255, 0, 0), 10)
        
plt.imshow(line_image)
```

```python
color_edges = np.dstack((edges, edges, edges))
combo = cv2.addWeighted(color_edges, 0.8, line_image, 1, 0)
plt.imshow(combo)
```

```python
final_combo = cv2.addWeighted(image, 0.8, line_image, 1, 0)
plt.imshow(final_combo)
```
